# Criando um equilibrio a partir do caos
### Gráficos gerados apartir da equação do <code>Logistic map</code>
    
<p> O logistic map é uma função criada por um biologo, Robert May, para simular o crescimento de uma população:</p>
    
<p><var>X<sub>(n+1)</sub><var> = <var>R * X <sub> n </sub></var> * (<var>1 - X<sub>n</sub></var>)<p/> 

<ul>
    <li> <var>X</var> = Valor atual</li>
    <li> <var>X<sub>n+1</sub></var> = Próxima iteração </li>
    <li> R = taxa de crescimento</li>
</ul>

<p> <strong>Código para plotar o gráfico através de uma função recursiva: </strong></p>

```python
import numpy as np
from matplotlib import pyplot as plt


def chaos(x, cont, val=2.5, array=np.zeros(2)):
    # formula
    x1 = val * x * (1 - x)
    
    # ajuste de contagem
    cont = cont + 1
    
    # gerando array para stack
    x = np.array([[x1, cont]])
    array = np.vstack((array, x))
    
    # quebra da recursividade
    if cont > 100:
        plt.plot(array[1:,1], array[1:, 0])
        plt.show()
        return array
    
    # gerando novos pontos atravez da recursividade
    else:
        return chaos(x1, cont, val, array)
    

array = np.zeros(2)
x1 = chaos(0.24, 0, 2.7)
```

## Mudando os valores da taxa de crescimento
<p>Alterando a taxa de crescimento para valores entre 0 e 1 obtem se um grãfico tendendo a zero, que para os estudos do biólogo (estudo de crescimento populacional) significaria a exinção da espécie, veja abaixo: </p>

```python
x2 = chaos(0.36, 0, 0.8)
```

<p> Ao aumentar a taxa de crescimento para valores superiores a <strong>3</strong> percebem-se uma repetição de padrões ao longo do gráfico</p>

```python
x3 = chaos(0.4, 0, 3)
```
